# 1. Filtrar por 1 o más condiciones

```python
# El uso de un filtro del DT según una condición en una variable

transport_sup_300 = employee[employee["transport_expense"]>300]
transport_sup_300

# transport_sup_300.shape
```

```python
# EL uso del isin 

pet_sup_2 = employee[employee["pet"].isin([3,4,5,6,7,8])]
pet_sup_2
```

```python
# el isin para var cuali

transportation_method = employee[employee["transportation_method"].isin(["bicycle","car"])]
transportation_method
```

```python
# Varios filtros var cuanti

dos_filtros = employee[(employee.transport_expense >= 250) & (employee.age <= 60)]
dos_filtros
```

```python
# Varios filtros var cuanti y cuali

dos_filtros_2 = employee[(employee.distance >= 20) & (employee.gender == 'M')]
dos_filtros_2
```

```python
# Uso del Query 1/2

querry1 = employee.query('transport_expense > 200 and gender == "F" ')

querry1
```

# 2. Odenar por 1 o más condiciones

DataFrame.sort_values(by, *, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)

```python
# Creamos una estructura de datos 

import numpy as np

df = pd.DataFrame({
     'col1': ['A', 'A', 'B', np.nan, 'D', 'C'],
     'col2': [2, 1, 9, 8, 7, 4],
     'col3': [0, 1, 9, 4, 2, 3],
     'col4': ['a', 'B', 'c', 'D', 'e', 'F']
 })

df
```

```python
# Ordenar por una columnna cuanti 

df.sort_values(by=['col2'])
```

```python
# Ordenar por una columnna cuali 

df.sort_values(by=['col1'])
```

```python
# Ordenar por dos columnas

df.sort_values(by=['col1', 'col2'])
```

```python
# Utilizar el ascenfing dentro de la función

df.sort_values(by='col1', ascending=False)
```

```python
# Sorting Data frames by multiple columns but different order

df.sort_values(by=['col1', 'col2'], ascending=[False, True])
```

```python
# Ordenar primero los NA

df.sort_values(by='col1', ascending=False, na_position='first')
```

# 3. Ver los distinct por 1 o varias variables 

pandas.unique(values)[source]

También se puede usar la función .drop_duplicates

```python
# El unique para una variable

pet = pd.DataFrame(employee.pet.unique())
pet.sort_values(by=0, ascending=True)
```

```python
# Obtener el número de valores únicos para una columna ---> función .nunique() , sin los NA

employee.pet.nunique(dropna=True)
```

```python
# Para dos columnas

import pandas as pd
import numpy as np
 
#Create a DataFrame
d = {
    'Name':['Alisa','Bobby','jodha','jack','raghu','Cathrine',
            'Alisa','Bobby','kumar','Alisa','Alex','Cathrine'],
    'Age':[26,24,23,22,23,24,26,24,22,23,24,24]
}
 
df = pd.DataFrame(d,columns=['Name','Age'])
df
```

```python
df.drop_duplicates() # Para crear los valores únicos para 2 o más variables
```

```python
# Hacer un valor un dataframe pero en sabe a una variable 

df = df.drop_duplicates(subset = ["Age"])
df
```
